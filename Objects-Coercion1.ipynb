R is a language that is-<br>
- Object Oriented
  Every item in R is an Object and has a type definition called Class. use .class() command to check<br>
  The same function can behave differently based on the class of the object.
<br><br>
- Functional Lanuage<br>
  Functions are first class objects in R
<br><br>
- Vectorized<br>
  R operations work on every element of the vector<br>
  Vector operations are preferred over loops
<br><br>
- Generic language<br>
  Type signatures do not need to be declared explicitly<br>
  functions use them to determine the object class
<br><br>
- Dynamic/Run time-<br>
  Only object values have types but the variables they are asigned to do not have types.
